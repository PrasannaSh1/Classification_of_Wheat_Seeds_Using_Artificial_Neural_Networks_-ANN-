Import modules for ANN 

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
import random
import numpy as np
```

Open the data set

```python
# Load the data set
def load_dataset():

    # open the data set
    file = open("seeds_dataset.txt")
    lines = file.readlines()
    data_set = []
    
    for line in lines:
        x = []
        line = line.strip().split()
        for elem in line:
            x.append(float(elem))
        data_set.append(x)
    
    return data_set

data_set = load_dataset()
```

```python
# for i in range(len(data_set)):
#     print(data_set[i])
```

```python
data_set = np.array(data_set)
```

```python
X = data_set[:, :-1]
y = data_set[:, -1]
```

```python
# print(y)
def split_data(X, y, split_ratio = 0.8):
    
    X, y = shuffle(X, y)  
    indexes = [i for i in range(len(X))]
    train_size = len(indexes) * split_ratio
    train_index = np.random.choice(indexes, size=int(train_size), replace=False)
    
    X_train = []
    y_train = []
    X_test = []
    y_test = []
    
    for i in range(len(X)):

        if i in train_index:
            X_train.append(X[i])
            y_train.append(y[i])
        
        else:
            X_test.append(X[i])
            y_test.append(y[i])
    
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
print(len(X_train), len(y_train))
```

ANN Specification 1

```python
# Create a MLPClassifier with 1 hidden layer of 32 neurons
model1 = MLPClassifier(hidden_layer_sizes=(32,), activation='logistic', solver='sgd', learning_rate_init=0.01, max_iter=200)
```

```python
# Train the model
model1.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred1 = model1.predict(X_test)
```

```python
# Compute accuracy
accuracy1 = accuracy_score(y_test, y_pred1)
print("Accuracy of ANN Specification 1: ", accuracy1*100)
```

```python
import matplotlib.pyplot as plt
train_scores = []
test_scores = []

# Train the model for 200 epochs
for i in range(200):
    model1.fit(X_train, y_train)
    train_scores.append(model1.score(X_train, y_train))
    test_scores.append(model1.score(X_test, y_test))
    if (i+1) % 10 == 0:
        print("Epoch:", i+1)
        print("Training Accuracy:", train_scores[-1])
        print("Test Accuracy:", test_scores[-1])

# Plot the training and test accuracy after every 10 epochs
plt.plot(train_scores, label='Training Accuracy')
plt.plot(test_scores, label='Test Accuracy')
plt.xlabel('Epochs (x10)')
plt.ylabel('Accuracy')
plt.title('Accuracy of MLP Classifier')
plt.legend()
plt.show()

# Print the final training and test accuracy
print("Final Training Accuracy:", train_scores[-1])
print("Final Test Accuracy:", test_scores[-1])
```

```python
# ANN Specification 2
# Create a MLPClassifier with 2 hidden layers of 64 and 32 neurons respectively
model2 = MLPClassifier(hidden_layer_sizes=(64, 32), activation='logistic', solver='sgd', learning_rate_init=0.01, max_iter=200)
```

```python
# Train the model
model2.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred2 = model2.predict(X_test)

# Compute accuracy
accuracy2 = accuracy_score(y_test, y_pred2)
print("Accuracy of ANN Specification 2: ", accuracy2*100)
```
